```python
def direct_polynom(x, arr):
  res = 0
  for i in range(len(arr)):
    value = 1
    for j in range(i):
      value *= x
    res += arr[i] * value
  return res
```

```python
def horner_method(x, arr):
  result = arr[0] 
  for i in range(1, len(arr)):
    result = result * x + arr[i]
  return result
```

```python
#horner's method time complexity calculation
average_arr = []
for n in range(1, 2001):
  average = 0
  arr = np.random.rand(n)
  for i in range(4):
    firstTs = time.time()
    horner_method(1.5, arr)
    secondTs = time.time()
    average += secondTs - firstTs
  average /= 5
  average_arr.append(average)
l_l = []
сoeff = average_arr[-1] / (2000)
for n in range(1, 2001):
  l_l.append(сoeff * n)
l = np.array(l_l)
y = np.arange(1, 2001)
x = np.array(average_arr)
plt.xlabel("N")
plt.ylabel("Time (seconds)")

plt.plot(y, x, color = "red", label = "empirical")
plt.plot(y, l, color = "blue", label = "theoretical")
plt.legend(title='Where:')

plt.show()
```

```python
def bubble_sort(array): 
    for i in range(0, len(array) - 1): 
        for j in range(len(array) - 1): 
            if (array[j] > array[j + 1]): 
                temp = array[j] 
                array[j] = array[j + 1] 
                array[j + 1] = temp 
    return array 
```

```python
#bubble sort time complexity calculation
average_arr = []
for n in range(1, 1001):
  average = 0
  arr = np.random.rand(n)
  for i in range(4):
    firstTs = time.time()
    bubble_sort(arr)
    secondTs = time.time()
    average += secondTs - firstTs
  average /= 5
  average_arr.append(average)
l_l = []
сoeff = average_arr[-1] / (1000 * 1000)
for n in range(1, 1001):
  l_l.append(сoeff * n * n)
l = np.array(l_l)
y = np.arange(1, 1001)
x = np.array(average_arr)
plt.xlabel("N")
plt.ylabel("Time (seconds)")

plt.plot(y, x, color = "red", label = "empirical")
plt.plot(y, l, color = "blue", label = "theoretical")
plt.legend(title='Where:')

plt.show()
```

```python
def quickSort(array):
  if len(array) <= 1:
        return array
  else:
    q = array[int(len(array) / 2)]
    s_nums = []
    m_nums = []
    e_nums = []
    for n in array:
        if n < q:
          s_nums.append(n)
        elif n > q:
          m_nums.append(n)
        else:
          e_nums.append(n)
    return quickSort(s_nums) + e_nums + quickSort(m_nums)
```

```python
#quicksort time complexity calculation
average_arr = []
for n in range(1, 2001):
  average = 0
  arr = np.random.rand(n)
  for i in range(4):
    firstTs = time.time()
    quickSort(arr)
    secondTs = time.time()
    average += secondTs - firstTs
  average /= 5
  average_arr.append(average)
l_l = []
сoeff = average_arr[-1] / (2000 * math.log(2000))
for n in range(1, 2001):
  l_l.append(сoeff * n * math.log(n))
l = np.array(l_l)
y = np.arange(1, 2001)
x = np.array(average_arr)
plt.xlabel("N")
plt.ylabel("Time (seconds)")

plt.plot(y, x, color = "red", label = "empirical")
plt.plot(y, l, color = "blue", label = "theoretical")
plt.legend(title='Where:')

plt.show()
```

```python
minrun = 32

def InsSort(arr, start, end):    
    for i in range(start + 1, end + 1):
        elem = arr[i]
        j = i - 1
        while j >= start and elem < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = elem
    return arr
def merge(arr, start, mid, end):
    if mid == end:
        return arr
    first = arr[start:mid + 1]
    last = arr[mid + 1:end + 1]
    len1 = mid - start + 1
    len2 = end - mid
    ind1 = 0
    ind2 = 0
    ind  = start
     
    while ind1 < len1 and ind2 < len2:
        if first[ind1] < last[ind2]:
            arr[ind] = first[ind1]
            ind1 += 1
        else:
            arr[ind] = last[ind2]
            ind2 += 1
        ind += 1
     
    while ind1 < len1:
        arr[ind] = first[ind1]
        ind1 += 1
        ind += 1
              
    while ind2 < len2:
        arr[ind] = last[ind2]
        ind2 += 1
        ind += 1   
              
    return arr
            
def timSort(arr):
    n = len(arr)
    
    for start in range(0, n, minrun):
        end = min(start + minrun - 1, n - 1)
        arr = InsSort(arr, start, end)
        
    curr_size = minrun
    while curr_size < n:    
        for start in range(0, n, curr_size * 2):
            mid = min(n - 1, start+  curr_size - 1)
            end = min(n - 1, mid + curr_size)
            arr = merge(arr, start, mid, end)
        curr_size *= 2
    return arr
```

```python
#timsort time complexity calculation
average_arr = []
for n in range(1, 2001):
  average = 0
  arr = np.random.rand(n)
  for i in range(4):
    firstTs = time.time()
    timSort(arr)
    secondTs = time.time()
    average += secondTs - firstTs
  average /= 5
  average_arr.append(average)
l_l = []
сoeff = average_arr[-1] / (2000 * math.log(2000))
for n in range(1, 2001):
  l_l.append(сoeff * n * math.log(n))
l = np.array(l_l)
y = np.arange(1, 2001)
x = np.array(average_arr)
plt.xlabel("N")
plt.ylabel("Time (seconds)")

plt.plot(y, x, color = "red", label = "empirical")
plt.plot(y, l, color = "blue", label = "theoretical")
plt.legend(title='Where:')

plt.show()
```
